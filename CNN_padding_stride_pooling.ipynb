#Padding

```python
model = Sequential()
#create three Convolution Layer
model.add(Conv2D(32,kernel_size=(3,3),padding = 'valid', activation = 'relu', input_shape = (28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding = 'valid', activation = 'relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding = 'valid', activation = 'relu'))
#convert img to 1D array
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```

```python
model = Sequential()
#create three Convolution Layer
model.add(Conv2D(32,kernel_size=(3,3),padding = 'same', activation = 'relu', input_shape = (28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding = 'same', activation = 'relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding = 'same', activation = 'relu'))
#convert img to 1D array
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```

#Stride

```python
model = Sequential()
#create three Convolution Layer
model.add(Conv2D(32,kernel_size=(3,3),strides = (1,1), activation = 'relu', input_shape = (28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),strides = (1,1), activation = 'relu'))
model.add(Conv2D(32,kernel_size=(3,3),strides = (1,1), activation = 'relu'))
#convert img to 1D array
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```

```python
model = Sequential()
#create three Convolution Layer
#stride: movement of kernel
model.add(Conv2D(32,kernel_size=(3,3),strides = (2,2), activation = 'relu', input_shape = (28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),strides = (2,2), activation = 'relu'))
model.add(Conv2D(32,kernel_size=(3,3),strides = (2,2), activation = 'relu'))
#convert img to 1D array
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

#Pooling

```python
from keras.layers import MaxPooling2D
#sride = (2,1)
model = Sequential()
#create three Convolution Layer
#stride: movement of kernel
model.add(Conv2D(32,kernel_size=(3,3),padding = 'valid', activation = 'relu', input_shape = (28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2),strides = 2,padding = 'valid' ))
model.add(Conv2D(32,kernel_size=(3,3),padding = 'valid', activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides = 2 ,padding = 'valid'))
#convert img to 1D array
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation = 'softmax'))
```

```python
model.summary()
```
